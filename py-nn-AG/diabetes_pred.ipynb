```python
import numpy as np
import pandas as pd
from sklearn.ensemble import RandomForestClassifier as rdf
import numpy as np
from sklearn.model_selection import train_test_split as tts
from matplotlib import pyplot as plt
import seaborn as sns
```

```python
data = pd.read_csv(r'/Users/abhraneelsaha/Desktop/diabetes.csv')
```

```python
data.head()
```

```python
data.shape
```

```python
y = [data['Outcome']]
X = []
for name in data:
    if name != 'Outcome':
        X += [list(data[name])]
        
X = np.array(X).transpose()
y = np.array(y).transpose()
        
x_train,x_test,y_train,y_test = tts(X, y, test_size=0.30, random_state=45)
```

```python
x_train
```

```python
y_train
```

```python
model = rdf()
model.fit(x_train,y_train)
print(model.score(x_train,y_train))
print(model.score(x_test,y_test))
```

```python
y_predictions = model.predict(x_test)

y_test_plot = y_test.transpose()[0]
y_predictions_plot = y_predictions.transpose()[0]
```

```python
sns.lineplot(y_test_plot,y_predictions_plot)
plt.show()
```
